## Sigmoid

```python
import math
```

```python
def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
sigmoid(200)
```

```python
sigmoid(1)
```

```python
sigmoid(-23)
```

```python
sigmoid(0.5)
```

## tanh

```python
def tanh(x):
  return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(30)
```

```python
tanh(-30)
```

```python
tanh(1)
```

```python
tanh(0.5)
```

## ReLU (Rectified Linear Unit)

```python
def relu(x):
  return max(0,x)
```

```python
relu(30)
```

```python
relu(-30)
```

```python
relu(1)
```

```python
relu(0.5)
```

## Leaky ReLU

```python
def leaky_relu(x):
  return max(0.1*x,x)
```

```python
leaky_relu(30)
```

```python
leaky_relu(-30)
```

```python
leaky_relu(1)
```

```python
leaky_relu(0.5)
```

```python
leaky_relu(-0.5)
```

## Activation Functions with PyTorch

```python
import torch
```

```python
print(torch.randn(4,4))
```

```python
input_layer = torch.tensor([[-0.1109,  0.2522,  0.1926, -2.3325]]) # torch.randn(1, 4)
weight_1 = torch.tensor([[-0.0486, -0.4529,  1.0933,  0.5512], [-0.8792,  1.1581,  1.7431, -0.9178], [ 0.0559, -0.5587,  2.7092,  0.3953], [ 0.0255,  0.7092,  0.0241, -0.3632]])
weight_2 = torch.randn(4, 4)
weight_3 = torch.randn(4, 4)
```

```python
hidden_1 = torch.matmul(input_layer, weight_1)
hidden_2 = torch.matmul(hidden_1, weight_2)

print(torch.matmul(hidden_2, weight_3))
```

```python
# Calculate weight_composed_1 and weight
weight_composed_1 = torch.matmul(weight_1, weight_2)
weight = torch.matmul(weight_composed_1, weight_3)

# Multiply input_layer with weight
print(torch.matmul(input_layer, weight))
```

Let's use ReLU

```python
import torch.nn as nn
```

```python
relu = nn.ReLU()

hidden_1_activated = relu(torch.matmul(input_layer, weight_1))
hidden_2_activated = relu(torch.matmul(hidden_1_activated, weight_2))
print(torch.matmul(hidden_2_activated, weight_3))

weight_composed_1_activated = relu(torch.matmul(weight_1, weight_2))

# Multiply `weight_composed_1_activated` with `weight_3
weight = torch.matmul(weight_composed_1_activated, weight_3)

# Multiply input_layer with weight
print(torch.matmul(input_layer, weight))
```
